```python
# necessary functions, run this cell only once!

# encoding of datetime features as seconds since epoch
# input: df which shall be encoded, dataframe time_feat containing only the values of datetime features
def dt_encoding(time_feat, df):
    for feat in time_feat:
        if feat in df.columns:
            df[feat] = time_feat[feat].values.astype('int64')// 10 ** 9
            df[feat].where(df[feat]>0, 0, inplace=True)
    return df

# decoding of datetime features encoded with the above method
# input: df which shall be decoded, dataframe time_feat containing only the values of datetime features
# negative values are decoded as nan
def dt_decoding(time_feat, df):
    for feat in time_feat:
        if feat in df.columns:
        # decode negative values as nans
            df[feat].where(df[feat] > 0, np.nan, inplace=True)
            df[feat] = pd.to_datetime(df.loc[:,feat]*10**9, unit='ns')
    return df

# deleting features
# input: df where feature should be droped, list of features to drop
def drop_feat(df, to_drop):
    for i in to_drop:
        if i in df.columns:
            df.drop(i, axis = 1, inplace=True)
    return df

# get binary features which should be droped because of too low variance
def drop_var(df):
    bin_features = df.select_dtypes(include=['number'])

    # get binary features
    for i in range(data.select_dtypes(include=['number']).shape[1]):
        if (len(np.unique(data.select_dtypes(include='number').iloc[:,i])) <= 2):
            continue
        else:
            bin_features = bin_features.drop(data.select_dtypes(include=['number']).columns[i], axis=1)
        
    # apply variance threshold
    sel = VarianceThreshold(threshold=(0.95*(1-0.95)))
    sel.fit_transform(bin_features)
    mask = sel.get_support()
    drop_var = []

    # get names of binary features with reasonable variance
    for i in range(len(bin_features.columns)):
        if not mask[i]: drop_var.append(bin_features.columns[i])
    return drop_var

# necessary variables
times = data.select_dtypes(include='datetime')
categorical = data.select_dtypes(include='object')
```

```python
tf.keras.backend.set_floatx('float64')
tf.random.set_seed(1)

# generate NN models
def generate_nn(u, a, optimizer='rmsprop', init='glorot_uniform'):
    model = Sequential()
    for i in range(len(u)):
        model.add(Dense(u[i], kernel_initializer=init, activation=a[i]))
    
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# generate validation set from training set
X_train_nn, X_test_nn, y_train_nn, y_test_nn = train_test_split(X_norm, y_enc, random_state=42)
X_train_nn, X_val_nn, y_train_nn, y_val_nn = train_test_split(X_train_nn, y_train_nn, test_size=0.2, random_state=1)


# gridSearch for NN
nn_models = []
# generate models and save all trained models
nn01 = KerasClassifier(build_fn=lambda: generate_nn([len(X_train_nn.columns), 256, 2], ['relu', 'relu', 'softmax'], 'adam'), epochs=10)
nn01.fit(X_train_nn, y_train_nn)
nn_models.append(nn01)

nn02 = KerasClassifier(build_fn=lambda: generate_nn([len(X_train_nn.columns), 256, 2], ['softmax', 'softmax', 'sigmoid'], 'adam'), epochs=10)
nn02.fit(X_train_nn, y_train_nn)
nn_models.append(nn02)

nn03 = KerasClassifier(build_fn=lambda: generate_nn([len(X_train.columns), 256, 2], ['relu', 'relu', 'sigmoid'], 'adam'), epochs=10)
nn03.fit(X_train_nn, y_train_nn)
nn_models.append(nn03)

nn04 = KerasClassifier(build_fn=lambda: generate_nn([len(X_train.columns), 256, 2], ['sigmoid', 'sigmoid', 'softmax'], 'adam'), epochs=10)
nn04.fit(X_train_nn, y_train_nn)
nn_models.append(nn04)
    
val_nn = pd.DataFrame(index=['Accuracy', 'Recall', 'Precision', 'F1', 'ROC_AUC', 'Hamming_loss'])
train_nn = pd.DataFrame(index=['Accuracy', 'Recall', 'Precision', 'F1', 'ROC_AUC', 'Hamming_loss'])
for m in nn_models:
    nn_pred = m.predict(X_val_nn)
    nn_train = m.predict(X_train_nn)
        
    # evaluation of trained models
    val_nn['NN_'+str(nn_models.index(m)+1)] = [accuracy_score(y_val_nn, nn_pred),\
                       recall_score(y_val_nn, nn_pred),\
                       precision_score(y_val_nn, nn_pred),\
                       f1_score(y_val_nn, nn_pred),\
                       roc_auc_score(y_val_nn, nn_pred),\
                       hamming_loss(y_val_nn,nn_pred)]
    train_nn['NN_'+str(nn_models.index(m)+1)] = [accuracy_score(y_train_nn, nn_train),\
                       recall_score(y_train_nn, nn_train),\
                       precision_score(y_train_nn, nn_train),\
                       f1_score(y_train_nn, nn_train),\
                       roc_auc_score(y_train_nn, nn_train),\
                       hamming_loss(y_train_nn, nn_train)]
val_nn = val_nn.T
train_nn = train_nn.T
```

```python
# functions for ranking of features
def topKfeat(score, K):
    cnt = 0
    printed = []
    # get names of features and rank
    ranking = []
    sorted_feat = score.sort_values(0, ascending=False)
    for feat in sorted_feat.Feature:
            if feat not in printed:
                ranking.append([feat, len(ranking)+1])
                printed.append(feat)
            if len(printed) > K-1:
                break
    return pd.DataFrame(data=ranking, columns=["Feature", "Rank"])

# ranking of feature with different models merged into one ranking
def model_rank(rankings):
    # get all used features
    features = list(rankings[0].Feature)
    for i in range(len(rankings)):
        for feat in list(rankings[i].Feature):
            if feat not in features:
                features.append(feat)

    # calculate rankings between all models
    score = pd.DataFrame([0]*len(features), index=features, columns=['Score'])
    for i in range(len(rankings)):
        for feat in rankings[i].Feature:
        # if already calculated, continue
            if score.Score.loc[feat] != 0:
                continue
        # if not, calculated score
            score_sum = 0
            for j in range(len(rankings)):
                if feat in list(rankings[j].Feature):
                    score_sum += rankings[j].loc[rankings[j].Feature == feat, 'Rank'].iloc[0]
                # penalty if feature is not in another ranking
                if feat not in list(rankings[j].Feature):
                    score_sum += len(features)
            score.loc[feat].Score=score_sum
            
    score['Score'].replace({0 : np.inf}, inplace=True)
    
    return score.sort_values('Score')

# input X, y get chi2 ranking regarding y
def chi2_rank(X_enc, y_enc):
    sel_chi2 = SelectKBest(chi2, k=len(X_enc.columns))
    sel_chi2.fit_transform(X_enc, y_enc)
    num_names = X_enc.columns.values[sel_chi2.get_support()]
    num_scores = sel_chi2.scores_[sel_chi2.get_support()]
    num_names_scores = list(zip(num_names, num_scores))
    chi2_df = pd.DataFrame(data = num_names_scores, columns=['Feature', 0])
    # get all ranked
    chi2_ranked = topKfeat(chi2_df, len(X_enc.columns))
    return chi2_ranked
   
# input X, y get f1 ranking regarding y
def f1_rank(X_enc, y_enc):
    sel_f_score = SelectKBest(f_classif, k=len(X_enc.columns))
    sel_f_score.fit_transform(X_enc, y_enc)
    num_names = X_enc.columns.values[sel_f_score.get_support()]
    num_scores = sel_f_score.scores_[sel_f_score.get_support()]
    num_names_scores = list(zip(num_names, num_scores))
    f_df = pd.DataFrame(data = num_names_scores, columns=['Feature', 0])
    # get all ranked
    f_ranked = topKfeat(f_df, len(X_enc.columns))
    return f_ranked

# input X, y get feature importance of each feature regarding y with untrained random forest rf
def rf_rank(X_train, y_train, rf):
    # train rf with the parameters decided above
    rf.fit(X_train,y_train)

    # get feature importance
    rf_feat_imp = pd.DataFrame(rf.feature_importances_)
    rf_feat_imp['Feature'] = X_train.columns

    return topKfeat(rf_feat_imp, len(X_train.columns))

# input X, y get permutated feature importance of each feature regarding y with untrained svc 
def svc_rank(X_train, y_train, svc):
    # train svc with parameters decided above
    svc.fit(X_train,y_train)
    # get feature permutation importances
    perm_imp_svc = permutation_importance(svc, X_train, y_train, random_state=42, n_jobs=-1, scoring='accuracy')
    # save as df
    svc_feat_imp = pd.DataFrame(X_train.columns, columns=['Feature'])
    svc_feat_imp[0] = perm_imp_svc.importances_mean
    
    return topKfeat(svc_feat_imp, len(X_train.columns))

# input X, y get permutated feature importance of each feature regarding y with nn
def nn_rank(X_train, y_train_ohe, nn):
    tf.random.set_seed(1)
    nn.fit(X_train, y_train_ohe)
    
    perm_nn = PermutationImportance(nn, random_state=42).fit(X_train,y_train_ohe)
    nn_feat_imp = eli5.explain_weights_df(perm_nn, feature_names=X_train.columns.tolist()).iloc[:,:-1]
    nn_feat_imp.columns = ['Feature', 0]
    return topKfeat(nn_feat_imp, len(X_train.columns))
```

```python
# generate feature ranking for each model
chi2_ranking = chi2_rank(X_enc, y_enc)
f1_ranking = f1_rank(X_enc, y_enc)
rf_ranking = rf_rank(X_train, y_train, rf)
svc_ranking = svc_rank(X_train_svc, y_train, svc)
nn_ranking = nn_rank(X_train_svc, y_train, nn)
```

```python
def all_rank(ranks, X_columns):
    rank_all = model_rank(ranks)
    rank_all['Feature'] = rank_all.index
    rank_all.index = [i for i in range(len(X_columns))]
# could be done with loop, but needs thinking about it
    all_rank = pd.DataFrame()
    all_rank['rank'] = ranks[0].Rank
    all_rank['chi2_feat'] = ranks[0].Feature
    all_rank['f_feat'] = ranks[1].Feature
    all_rank['rf_feat'] = ranks[2].Feature
    all_rank['svc_feat'] = ranks[3].Feature
    all_rank['nn_feat'] = ranks[4].Feature
    all_rank['final_score'] = rank_all.Score
    all_rank['final_feat'] = rank_all.Feature
    return all_rank

def metrics(y_true, y_pred):
    metrics = [accuracy_score(y_true, y_pred), \
               recall_score(y_true, y_pred), \
               precision_score(y_true, y_pred), \
               f1_score(y_true, y_pred), \
               roc_auc_score(y_true, y_pred),\
              hamming_loss(y_true, y_pred)]
    return metrics
```

```python
# generate final ranking with all rankings
all_ranking = all_rank([chi2_ranking, f1_ranking, rf_ranking, svc_ranking, nn_ranking], X_train.columns)
```

```python
# generate evaluation dataframe
evaluation = pd.DataFrame(index=['Accuracy','Recall', 'Precision', 'F1', 'ROC_AUC', 'Loss'])

np.random.seed(42)
tf.random.set_seed(42)

# train and test models with top 2+3*k features for k in range [0,12]
for j in range(2,41,3):
    i = all_ranking.iloc[:j, -1:]
    # random forest
    rf.fit(X_train[i.final_feat], y_train)
    y_pred_rf = rf.predict(X_test[i.final_feat])
    evaluation['RF '+str(len(i.final_feat))] = metrics(y_test, y_pred_rf)
    
    # svc
    svc.fit(X_train_svc[i.final_feat], y_train)
    y_pred_svc = svc.predict(X_test_svc[i.final_feat])
    evaluation['SVC '+str(len(i.final_feat))] = metrics(y_test, y_pred_svc)
    
    #nn
    nn.fit(X_train_svc[i.final_feat], y_train)
    y_pred_nn = nn.predict(X_test_svc[i.final_feat])
    evaluation['NN '+str(len(i.final_feat))] = metrics(y_test, y_pred_nn)
```
